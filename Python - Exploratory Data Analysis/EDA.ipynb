## EDA - Exploratory Data Analysis

```python
sns.countplot(y='Gender',data=df)
```

```python
cnt=sns.countplot(y='Gender',data=df)
for bars in cnt.containers:
    cnt.bar_label(bars)
```

```python
df.head()
```

```python
group=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=group)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
group=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=group)
```

### Female purchase more than male and purchasing power of females are also higher

```python
ageana=sns.countplot(x='Age Group', hue='Gender',data=df)
for ana in ageana.containers:
    ageana.bar_label(ana)
```

```python
Age_group=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Age_group)
```

### Most buyers are in age 26-35 and female

## State Analysis

```python
stat_sale=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=stat_sale)
```

```python
stat_amt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=stat_amt)
```

* Max orders and amount spent from uttar Pradesh*

## Marital Status

```python
Mars=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for ms in Mars.containers:
    Mars.bar_label(ms)
```

```python
mars_sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Marital_Status',y='Amount',data=mars_sales,hue='Gender')
```

### higher Purchase power in married females

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
Ox=sns.countplot(x='Occupation',data=df,hue='Occupation')
for O in Ox.containers:
    Ox.bar_label(O)
```

```python
Occ_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Occupation',y='Amount',data=Occ_sales,hue='Occupation')
```

Max buyers are from IT, Healthcare and Aviation sector

## Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
Pc=sns.countplot(x='Product_Category',data=df,hue='Product_Category')
for O in Pc.containers:
    Ox.bar_label(O)
```

```python
Pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_Category',y='Amount',data=Pc,hue='Product_Category')
```

Most Sale is food, clothing and electronics

```python
Pc=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_ID',y='Amount',data=Pc,hue='Product_ID')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
